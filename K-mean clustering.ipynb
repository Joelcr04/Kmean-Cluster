```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
make_blobs
```

```python
plt.rcParams['figure.figsize'] = (16, 9)

# Creating a sample dataset with 4 clusters
X, y = make_blobs(n_samples=100, n_features=3, centers=1)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

```python
# Initializing KMeans
kmeans1 = KMeans(n_clusters=3)
```

```python
# Fitting with inputs
kmeans1 = kmeans1.fit(X)
```

```python
# Predicting the clusters
labels = kmeans1.predict(X)
```

```python
labels

# Getting the cluster centers
Centroidarray = kmeans1.cluster_centers_

Centroidarray
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c='g')
ax.scatter(Centroidarray[:, 0], Centroidarray[:, 1], \
           Centroidarray[:, 2], marker='*', c='b', s=2000)
```
